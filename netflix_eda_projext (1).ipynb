```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mymoviedb.csv', engine='python')
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df['Release_Date'].unique()
```

```python
df['Release_Date']=pd.to_datetime(df['Release_Date'],errors='coerce')
```

```python
df.info()
```

```python
df['Release_year']=df['Release_Date'].dt.year
df['Release_month']=df['Release_Date'].dt.month
df['Release_Day']=df['Release_Date'].dt.day
```

```python
df.info()
```

```python
df=df.drop('Release_Date',axis=1,inplace=True)
```

```python
df
```

```python
col=['Overview','Original_Language','Poster_Url']
df.drop(col,axis=1,inplace=True)
```

```python
df
```

```python
df[df['Title'].isnull()]
```

```python
df.dropna(subset=['Title'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['Genre'].isnull()]
```

```python
df.dropna(subset=['Genre'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Genre']=df['Genre'].str.split(', ')
df=df.explode('Genre').reset_index(drop=True)
```

```python
df
```

```python
df.shape
```

```python
df['Genre'].value_counts().plot.pie(y=df['Genre'],figsize=(25,16),autopct='%1.1f')
```

drama is the most common genere

```python
df['Genre'].value_counts().plot.bar(figsize=(15,16))
```

```python
df.info()
```

```python
df['Vote_Count']=df['Vote_Count'].astype(int)
```

```python
df.info()
```

```python
genre_votes=df.groupby('Genre')['Vote_Count'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_votes.head(10).values, y=genre_votes.head(10).index)
plt.title('Top 10 Genres by Total Vote Count')
plt.xlabel('Total Votes')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()
```

drama has maximum vote count
